```python
import pandas as pd

# Load dataset
df = pd.read_csv("final_products (1).csv")
print("Original rows:", len(df))






```

```python
import mysql.connector

conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="$Abirami",
    database="webscraper_produits"
)

cursor = conn.cursor()
print("Connected Successfully")
```

```python
import os
os.listdir()
```

```python
cursor.execute("DROP TABLE IF EXISTS final_products")

cursor.execute("""
CREATE TABLE final_products (
    id INT AUTO_INCREMENT PRIMARY KEY,
    category VARCHAR(100),
    product_name VARCHAR(255),
    brand VARCHAR(100),
    price DOUBLE,
    description TEXT,
    reviews INT,
    price_range VARCHAR(20),
    discount_flag VARCHAR(10),
    category_weight INT,
    color VARCHAR(50),
    is_expensive VARCHAR(10)
)
""")

conn.commit()
print("Table Created Successfully")
```

```python
for _, row in df.iterrows():
    cursor.execute("""
        INSERT INTO final_products
        (category, product_name, brand, price, description, reviews, 
         price_range, discount_flag, category_weight, color, is_expensive)
        VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
    """, tuple(row))

conn.commit()
print("All Data Inserted Successfully")
```

```python
cursor.execute("SELECT COUNT(*) FROM final_products")
print("Total Records in MySQL:", cursor.fetchone())
```
